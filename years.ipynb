```python
import pandas as pd
# import csv files
adult21 = pd.read_csv('adult21.csv')
adult20 = pd.read_csv('adult20.csv')
adult19 = pd.read_csv('adult19.csv')
```

```python

att21 = ['SEX_A','REGION','WEIGHTLBTC_A','DEPFREQ_A','ANXFREQ_A','PAIFRQ3M_A','PHSTAT_A', 'EDUCP_A','CANEV_A', 'AGEP_A']
att20 = ['SEX_A','REGION','WEIGHTLBTC_A','DEPFREQ_A','ANXFREQ_A','PAIFRQ3M_A','PHSTAT_A', 'EDUC_A','CANEV_A', 'AGEP_A']
att19 = ['SEX_A','REGION','WEIGHTLBTC_A','DEPFREQ_A','ANXFREQ_A','PAIFRQ3M_A','PHSTAT_A', 'EDUC_A','CANEV_A', 'AGEP_A']

data_21 = adult21[att21]
data_20 = adult20[att20]
data_19 = adult19[att19]

# Remove columns where 'SEX_A' isn't a 1 or 2
data_drop_21 = data_21[(data_21['SEX_A'] == 7) | (data_21['SEX_A'] == 9)].index
data_21 = data_21.drop(data_drop_21, inplace = False)

data_drop_20 = data_20[(data_20['SEX_A'] == 7) | (data_20['SEX_A'] == 9)].index
data_20 = data_20.drop(data_drop_20, inplace = False)

data_drop_19 = data_19[(data_19['SEX_A'] == 7) | (data_19['SEX_A'] == 9)].index
data_19 = data_19.drop(data_drop_19, inplace = False)


# data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# filter PAIFRQ3M_A = 1 or 2 or 3 or 4 for data_21, data_20, data_19
data_21 = data_21[(data_21['PAIFRQ3M_A'] >= 1) & (data_21['PAIFRQ3M_A'] <= 4)]
data_20 = data_20[(data_20['PAIFRQ3M_A'] >= 1) & (data_20['PAIFRQ3M_A'] <= 4)]
data_19 = data_19[(data_19['PAIFRQ3M_A'] >= 1) & (data_19['PAIFRQ3M_A'] <= 4)]


# filter DEPFREQ_A = 1:5 for data_21, data_20, data_19
data_21 = data_21[(data_21['DEPFREQ_A'] >= 1) & (data_21['DEPFREQ_A'] <= 5)]
data_20 = data_20[(data_20['DEPFREQ_A'] >= 1) & (data_20['DEPFREQ_A'] <= 5)]
data_19 = data_19[(data_19['DEPFREQ_A'] >= 1) & (data_19['DEPFREQ_A'] <= 5)]

# plot REPSTRAIN vs DEPFREQ_A for data_21, data_20, data_19

# make figure
fig, ax = plt.subplots(1, 3, figsize=(15, 7))

fig.suptitle('Pain v. Depression Frequency: By Year', fontsize=30, fontweight='bold', fontstyle = 'italic')

# subplot 1, hide xlabel
plt.subplot(1,3,1)
sns.countplot(x='PAIFRQ3M_A', hue='DEPFREQ_A', data=data_19)
plt.title('2019')
plt.xticks([0,1,2,3], ['Never', 'Sometimes', 'Often', 'Everyday'])
plt.xlabel('')
plt.ylabel('Depression Response Count')
plt.ylim(0, 9000)

plt.legend(['Daily', 'Weekly', 'Monthly', 'A few times a year', 'Never'], loc='upper left', bbox_to_anchor=(-0.5, 1.0), title = 'Depression Frequency')

# subplot 2
plt.subplot(1,3,2)
sns.countplot(x='PAIFRQ3M_A', hue='DEPFREQ_A', data=data_20)
plt.title('2020')
plt.xticks([0,1,2,3], ['Never', 'Sometimes', 'Often', 'Everyday'])
plt.xlabel('')
plt.ylabel('')
plt.ylim(0, 9000)
plt.legend().set_visible(False)



# subplot 3
plt.subplot(1,3,3)
sns.countplot(x='PAIFRQ3M_A', hue='DEPFREQ_A', data=data_21)
plt.title('2021')
plt.xticks([0,1,2,3], ['Never', 'Sometimes', 'Often', 'Everyday'])
plt.xlabel('')
plt.ylabel('')
plt.ylim(0, 9000)
plt.legend().set_visible(False)

# format figure
plt.tight_layout()

plt.show()





```
